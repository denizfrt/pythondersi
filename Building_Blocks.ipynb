## Dize İndeksleme
Python'da stringler, Unicode karakterlerini temsil eden bayt dizileridir. Bu, Python'un yerleşik bir string indeksine sahip olduğu anlamına gelir. String indekslemesi `0`'dan başlayarak `n-1`'e kadar devam eder; burada `n`, string uzunluğudur. Bazı örnekler görelim.

```python
# String indexing examples
hello_variable = 'Hello, World!'

# Get the character at position 1 (remember that the first character has the position 0)
print(hello_variable[1])

# Get the characters from position 2 to position 5 (not included)
print(hello_variable[2:5])

# We can also work backwards using the `-` sign. So -1
# will give us the last letter, -2 will give us the second last etc.
print(hello_variable[-2])
```


## Kendi Fonksiyonunuzu Yazın
Artık stringler(dizgi), sayılar ve değişkenler hakkında bilgi sahibi olduğunuza göre, bu bilgilerinizi bir fonksiyonla test edelim. Bir dizgiyi girdi olarak alan ve bu dizginin ilk harfini küçük harfle, son harfini ise büyük harfle döndüren bir fonksiyon yazmaya çalışın. Örneğin, girdi dizgisi 'Hello, World!' ise, çıktı 'hD' olmalıdır.


Unutmayın: fonksiyonlar genellikle bir girdi alır ve bir işlem yaptıktan sonra bir çıktı döndürür. Her zaman `def` ile başlar, ardından açık bir parantez ve fonksiyonunuzun alacağı değerin adı (esas olarak bir yer tutucu) gelir, ardından iki nokta üst üste gelir. Aşağıdaki satırda/satırlarda girdi üzerinde bir işlem yapar, genellikle yeni bir değişkene atar ve ardından bu değişkeni döndürür.

```python
# Giriş kelimesini alan ve ilk harfini sonuna koyan, bu kelimenin yeni bir versiyonunu
# new_word adlı yeni bir geçici değişkene atayan ve ardından bu new_word'ü döndüren örnek bir işlev
def letter_swap(word):
    new_word = word[1:] + word[0]
    return new_word
```
