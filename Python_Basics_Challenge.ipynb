## Create conditions


20'ye eşit yeni bir değişken `e` oluşturun ve bu değişken gerçekten 5'ten büyükse `"5'ten büyük"` ifadesini gösterecek bir koşul oluşturun.




💡 Python'da bir koşulu doğru şekilde nasıl yazacağınızı öğrenmek için dersi inceleyin.

```python
e =20;
if e>5:print("5'ten büyük")
```

Değişken 5'ten büyük değilse `"5'ten küçük veya 5'e eşit"` ifadesini görüntülemek için önceki kodu tamamlayın.




`e` değişkenine farklı değerler atamayı deneyin ve kodunuzu bu yeni değerlerle test edin.

```python
e =3;
if e>5:print("5'ten büyük")
else:print("5'ten küçük veya 5'e eşit")
```

e'nin 5'e eşit olması durumunda `"5'e eşit"` ifadesini görüntülemek için önceki kodu tamamlayın.


Kodunuzu birkaç farklı `e` değeriyle test edin.


<details>
<summary>İpucu 💡</summary>


`elif`
</details>

```python
e =5;
if e>5:print("5'ten büyük")
elif e==5:print("5'e eşit")
else:print("5'ten küçük")
```


Bir döngü kullanarak, `list_example` içindeki 4'ün katlarını birbirinin altına yazdırın.


<details>
<summary>İpucu 💡</summary>


`val` değişkeninin 4'ün katı olup olmadığını test etmek için aşağıdaki koşulu yazabilirsiniz: `val % 4 == 0`.



Python'da `%` işareti [modulo](https://simple.wikipedia.org/wiki/Modulation_(matematik)) anlamına gelir. `val`'ın 4'e bölünmesinin kalanının 0 olup olmadığını test ediyoruz.
</details>

```python
for i in list_example:
  if i%4==0:
    print(i)

```

## Dictionaries


Sözlüğü başlatmak için aşağıdaki kodu çalıştırın.

```python
dict_example = {
    "A": "car",
    "B": 2,
    "C": "bike"
}
print(dict_example)
```

`dict_example` anahtarlarını görüntüle

```python
print(dict_example.keys())
```

`C` key'i ile ilişkili değeri görüntüle

```python
print(dict_example["C"])
```

`8` değerine sahip yeni bir `D` anahtarı ekleyin. Sonucu kontrol etmek için `dict_example`'ı görüntüleyin.

```python
dict_example = {
    "A": "car",
    "B": 2,
    "C": "bike",
    "D":8
}
print(dict_example)
```
